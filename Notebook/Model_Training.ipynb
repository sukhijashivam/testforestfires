```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv ')
```

```python
df.head()
```

```python
df.columns
```

```python
#3 Drop day,month and year
df.drop(['day', 'month', 'year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding

df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Independent and Dependent Features

X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
y
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

## Feature Selection based on correlation


```python
X_train.corr()
```

```python
## Check for multicollinearity

plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,thresold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > thresold:
                colname= corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

```python
# Thresold-Domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## Drop these features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

## Boxplots to understand the effects of Standard Scaler

```python

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train before scaling')

plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train after scaling')

plt.tight_layout()
plt.show()
```

### Model Training
## 1. Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
plt.scatter(y_test,y_pred)
```

## 2. Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
plt.scatter(y_test,y_pred)
```

## Cross Validation (Hypertuning)

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```

```python
lassocv.alpha_
```

```python
lassocv.alphas_
```

```python
lassocv.mse_path_
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
```

## 3. Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
plt.scatter(y_test,y_pred)
```

## Ridge Cross Validation

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
plt.scatter(y_test,y_pred)
```

```python
ridgecv.get_params()
```

## 4. Elastic Net (Ridge+Lasso)

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
enet=ElasticNet()
enet.fit(X_train_scaled,y_train)
y_pred=enet.predict(X_test_scaled)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
plt.scatter(y_test,y_pred)
```

## Elastic Net Cross Validation



```python
from sklearn.linear_model import ElasticNetCV
enetcv=ElasticNetCV(cv=5)
enetcv.fit(X_train_scaled,y_train)
y_pred=enetcv.predict(X_test_scaled)
mea=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error",mea)
print("Score : ",score)
plt.scatter(y_test,y_pred)
```
