```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.info()
```

## Data Cleaning


```python
dataset.isnull().sum()
```

```python
## Missing values

dataset[dataset.isnull().any(axis=1)]
```

The dataset is divided into two subsets based on the Region starting from the 122nd index. We can create a new column to indicate the Region for each entry:

Rows 0 to 121: "Bejaia Region Dataset"

Rows 122 onwards: "Sidi-Bel Abbes Region Dataset"

A new column named "Region" can be added to reflect this classification.



```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
## Removing null values

df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
#Remove 122th row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
# There are spaces in the columns name , we have to correct it
```

```python
## Fix spaces in column name

df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the required columns as integer data type

```python
df[['month','day','year','Temperature','RH','WS']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

## Changes the other columns as float data type

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
## Lets save the clean dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
